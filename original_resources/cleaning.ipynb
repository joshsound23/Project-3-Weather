```python
import pandas as pd
```

```python
# Step 1: Read the Parquet file into a pandas DataFrame
parquet_file = 'daily_weather.parquet'  # Replace with your actual Parquet file path
df = pd.read_parquet(parquet_file)
```

```python
df
```

```python
# Reducing columns

df = df[['station_id', 'city_name', 'date', 'season', 'avg_temp_c', 'precipitation_mm', 'snow_depth_mm', 'avg_wind_speed_kmh', 'sunshine_total_min']]
df
```

```python
# Reducing timeframe to 5 year history
df = df[df['date']>= '2018-01-01']
```

```python
df = df[df['date']< '2023-01-01']
```

```python
df
```
